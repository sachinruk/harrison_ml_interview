```python
import ast
import pandas as pd

df = pd.read_csv("./data/cats_and_dogs/pets_dataset_info.csv")
df["Breed"] = df["Breed"].apply(ast.literal_eval)
df.head()
```

```python
df.shape
```

To see how many times we see two pets:

```python
df["Breed"].map(len).value_counts()
```

```python
from collections import Counter
# Flatten the list of all breeds
all_breeds = [breed for sublist in df['Breed'] for breed in sublist]
# Count occurrences
breed_counts = Counter(all_breeds)
print(f"There are {len(breed_counts)} unique breeds in the dataset.")
breed_counts.most_common()
```

```python
import pathlib
from PIL import Image

df["image_path"] = df["Sample_ID"].map(lambda x: pathlib.Path(f"./data/cats_and_dogs/data/{x}/image.jpg"))
df["mask_path"] = df["Sample_ID"].map(lambda x: pathlib.Path(f"./data/cats_and_dogs/data/{x}/mask.jpg"))
df["image_size"] = df["image_path"].map(lambda x: Image.open(x).size)
df["mask_size"] = df["mask_path"].map(lambda x: Image.open(x).size)
df.sample(10)
```
